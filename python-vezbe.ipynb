# Pisanje programskih petlji

Automatizaciju izvrÅ¡avanja seta instrukcija moÅ¾emo ostvariti koriÅ¡Ä‡enjem programskih petlji: `for` i `while`

`for` - petlja koja izvrÅ¡ava naredbe preodreÄ‘eni broj puta
```
for i in range(a,b,korak):
  izvrsi komandu
```
petlja se izvrÅ¡ava za vrednosti `i` koje su u intervalu `[a,b)` tj. od celobrojnih `a` do `b` sa korakom `korak`.

`in` operator u ovom sluÄaju poziva upisivanje pojedinaÄnih vrednosti u promenljivu `i`.

````python
for i in range(1,11,2):
  print(i)
````

Funkcija `range` se moÅ¾e pozvati i sa jednim parametrom:
```
for i in range(n):
  izvrsi komandu
```
gde bi se u tom sluÄaju vrednosti `i` menjale od `0` do `n-1`.


````python
zbir = 0
for i in range(3):
  zbir += i

print(zbir)
````

`while` - petlja koja izvrÅ¡ava naredbe dokle god je difinisan logiÄki uslov uz nju.
```
while logicki_uslov:
  izvrsi komandu
```

Dodatna kontrola toka izvrÅ¡avanja u okviru petlji:
*   `break` - prekida izvrÅ¡avanje i izlazi iz unutraÅ¡nje petlje
*   `continue` - preskaÄe trenutnu iteraciju i nastavlja sa sledeÄ‡om
*   `else` - set instrukcija se izvrÅ¡ava nakon pravilnog zavrÅ¡etka petlje (npr. prolazak kroz sve elemente u `for`)

````python
broj = 0

while broj < 10:
  broj += 1

  if broj % 2 == 0:
    print('{} je paran broj'.format(broj))
    continue
  else:
    print('{} je neparan broj'.format(broj))
    continue

  print('naknadni ispis')
else:
  print('ZavrÅ¡ena petlja')
````

## List comprehension

Za kreiranje nove liste koristeÄ‡i elemente stare moÅ¾e se koristiti tzv. [*list comprehension*](https://www.w3schools.com/python/python_lists_comprehension.asp) koja predstavlja naÄin koriÅ¡Ä‡enja kraÄ‡e sintakse za pisanje `for` programske petlje.

Primer: Posmatrati listu celobrojnih vrednosti u intervalu `[1,10]`. Kreirati novu listu tako da su njeni elementi kvadrati parnih vrednosti prve liste.

````python
even_squares = []
for number in range(1,11):
  if number % 2 == 0:
    even_squares.append(number**2)

print(even_squares)
````

````python
even_squares = [number**2 for number in range(1,11) if number % 2 == 0]
print(even_squares)
````

````python
# smatracemo da su liste a i b jednakih dimenzija
def saberi_liste(a,b):
  c = []
  for ai,bi in zip(a,b):
    c.append(ai+bi)

  return c
````

4. Skalarno, vektorski i element po element pomnoziti v1 i v2.

````python
v1=np.array([[1],[1],[2],[1]])
v2=np.array([[2],[2],[0],[1]])

print(np.dot(v1.T,v2)) #skalarno: vrsta*kolona
print(np.dot(v1,v2.T)) #vektorski: kolona*vrsta
print(np.multiply(v1,v2)) #element po element
print(v1*v2) #element po element
````

7. IzraÄunati l1, l2 i l beskonacno normu vektora v1

````python
v1=np.array([[1],[1],[2],[1]])
print(v1)
print(np.linalg.norm(v1))
print(np.linalg.norm(v1,1))
print(np.linalg.norm(v1,2))
print(np.linalg.norm(v1,np.inf))
````

9. IzraÄunati rang matrice A i B

````python
A=np.array([[1,0,5],[2,6,1],[3,2,0]])
B=np.array([[1,2,3],[0,2,2],[4,0,4]])
print(np.linalg.matrix_rank(A))
print(np.linalg.matrix_rank(B))
````

Cilj veÅ¾be je da se studenti upoznaju sa osnovnim funkcionalnostima [matplotlib](https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html) paketa.

U nastavku Ä‡e biti prikazano nekoliko funkcija pomoÄ‡u kojih se kreiraju grafici koje Ä‡emo u daljem radu najÄeÅ¡Ä‡e koristiti.

````python
# uÄitavanje neophodnih biblioteka
import numpy as np
import matplotlib.pyplot as plt
````

````python
# kreiranje skupa podataka (dve varijable)
x=np.linspace(1,50,100) # linspace(start,end,num_points)
y=np.sin(x)
````

````python
#linijski grafik:
plt.plot(x,y)
````

````python
#grafik rasipanja
plt.scatter(x,y)
````

````python
#grafik rasipanja drugi naÄin
plt.plot(x,y,'x')
````

````python
#dva grafika u okviru istog prozora
plt.plot(x,y)
plt.scatter(x,y)
````

````python
# dva grafika u dva zasebna prozora
plt.plot(x,y)
fig=plt.figure()
plt.scatter(x,y)
````

````python
# nekoliko vizuelnih detalja na graficima
plt.plot(x,y, 'x-', color='red', alpha=.5, label='the data')

plt.title('The Title')
plt.xlabel('x axis label')
plt.ylabel('y axis label')

plt.legend(loc='best') #loc = 'best' makes the legend not overlap the data

plt.grid()

plt.savefig('test_fig.png')
````

````python
# linijski grafici i histogram
t=np.linspace(1,100,100)
x1=np.sin(20./(2*np.pi) * t)
x2=np.sin(40./(2*np.pi)*t)
x3=x1+x2

plt.figure()
plt.subplot(2,1,1)
plt.plot(t,x1,label='x1')
plt.plot(t,x2,label='x2')
plt.plot(t,x3,label='x3')
plt.legend(loc='best')

plt.subplot(2,1,2)
x1hist=plt.hist(x1,density='true', bins=25, alpha=.25,label='x1')
x1hist=plt.hist(x2,density='true', bins=25, alpha=.25,label='x2')
plt.legend(loc='best')
````
