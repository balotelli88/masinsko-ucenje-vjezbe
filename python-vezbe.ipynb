# Pisanje programskih petlji

Automatizaciju izvršavanja seta instrukcija možemo ostvariti korišćenjem programskih petlji: `for` i `while`

`for` - petlja koja izvršava naredbe preodređeni broj puta
```
for i in range(a,b,korak):
  izvrsi komandu
```
petlja se izvršava za vrednosti `i` koje su u intervalu `[a,b)` tj. od celobrojnih `a` do `b` sa korakom `korak`.

`in` operator u ovom slučaju poziva upisivanje pojedinačnih vrednosti u promenljivu `i`.

````python
for i in range(1,11,2):
  print(i)
````

Funkcija `range` se može pozvati i sa jednim parametrom:
```
for i in range(n):
  izvrsi komandu
```
gde bi se u tom slučaju vrednosti `i` menjale od `0` do `n-1`.


````python
zbir = 0
for i in range(3):
  zbir += i

print(zbir)
````

`while` - petlja koja izvršava naredbe dokle god je difinisan logički uslov uz nju.
```
while logicki_uslov:
  izvrsi komandu
```

Dodatna kontrola toka izvršavanja u okviru petlji:
*   `break` - prekida izvršavanje i izlazi iz unutrašnje petlje
*   `continue` - preskače trenutnu iteraciju i nastavlja sa sledećom
*   `else` - set instrukcija se izvršava nakon pravilnog završetka petlje (npr. prolazak kroz sve elemente u `for`)

````python
broj = 0

while broj < 10:
  broj += 1

  if broj % 2 == 0:
    print('{} je paran broj'.format(broj))
    continue
  else:
    print('{} je neparan broj'.format(broj))
    continue

  print('naknadni ispis')
else:
  print('Završena petlja')
````

## Petlje nad listama

`for` petlja se često koristi za prolazak po elementima nekih nizova u redosledu po kome se javljaju.

Osnovna varijanta posmatra indekse pojedinačnih elemenata po kojima se vrši pristup.
```
for idx in range(len(array)):
  item = lista[idx]
  print(item)
```

````python
lista = ['a','b',3]

for idx in range(len(lista)):
  item = lista[idx]
  print(item)
````

Alternativno, češći način pristupa je direktno postavljanje iteratora da uzima elemente niza.
```
for item in array:
  print(item)
```
gde su sada vrednosti `i` elementi ove liste.

````python
for item in ['a','b',3]:
  print(item)
````

## List comprehension

Za kreiranje nove liste koristeći elemente stare može se koristiti tzv. [*list comprehension*](https://www.w3schools.com/python/python_lists_comprehension.asp) koja predstavlja način korišćenja kraće sintakse za pisanje `for` programske petlje.

Primer: Posmatrati listu celobrojnih vrednosti u intervalu `[1,10]`. Kreirati novu listu tako da su njeni elementi kvadrati parnih vrednosti prve liste.

````python
even_squares = []
for number in range(1,11):
  if number % 2 == 0:
    even_squares.append(number**2)

print(even_squares)
````

````python
even_squares = [number**2 for number in range(1,11) if number % 2 == 0]
print(even_squares)
````

Funkcije predstavljaju imenovane sekvence instrukcija koje se izvršavaju po pozivu tog imena.

Za navedene osobine i operacije nad listama, postavlja se problem kako bi npr. sabrali elemente 2 liste koje sadrže brojčane vrednosti. Pošto operator `+` definiše spajanje listi, potrebno je definisati zasebnu funkciju koja bi rešila ovaj problem

````python
a = [1,2,3]
b = [4,5,6]
for ai, bi in zip(a,b):
    print(ai,bi)
````

````python
# smatracemo da su liste a i b jednakih dimenzija
def saberi_liste(a,b):
  c = []
  for ai,bi in zip(a,b):
    c.append(ai+bi)

  return c
````

````python
a = [1,2,10]
b = [4,5,6]
c = saberi_liste(a,b)
print(c)
````

# Indeksiranje

Izdvajanje pojedinačnih elemenata u okviru numpy niza je identično kao i za liste. Dodatno, funkcionalnost list slicing-a je takođe prisutna.

````python
a = np.array([1,2,3])

print(a[1])
print(a[:2])
````

Procedura u okviru 2D nizova je slična, gde je sada pritom potrebno indeksirati po svakoj dimenziji. Indeksi prve dimenzije predstavljaju vrste, dok druge dimenzije kolone.

````python
a = np.array([[1,2,3], [4,5,6]])
print(a)

print(a[0,[1,2]])
print(a[:,2])
````

Za potrebe izdvajanja elemenata često se koristi i logičko indeksiranje, u vidu kreiranja maski.

````python
a = np.array([[1,2,3], [4,5,6]])
mask = a % 2 == 0

print(a)
print(mask)
print(a[mask])
````

Cilj vežbe je da se studenti upoznaju sa osnovnim funkcionalnostima [matplotlib](https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html) paketa.

U nastavku će biti prikazano nekoliko funkcija pomoću kojih se kreiraju grafici koje ćemo u daljem radu najčešće koristiti.

````python
# učitavanje neophodnih biblioteka
import numpy as np
import matplotlib.pyplot as plt
````

````python
# kreiranje skupa podataka (dve varijable)
x=np.linspace(1,50,100) # linspace(start,end,num_points)
y=np.sin(x)
````

````python
#linijski grafik:
plt.plot(x,y)
````

````python
#grafik rasipanja
plt.scatter(x,y)
````

````python
#grafik rasipanja drugi način
plt.plot(x,y,'x')
````

````python
#dva grafika u okviru istog prozora
plt.plot(x,y)
plt.scatter(x,y)
````

````python
# dva grafika u dva zasebna prozora
plt.plot(x,y)
fig=plt.figure()
plt.scatter(x,y)
````

````python
# nekoliko vizuelnih detalja na graficima
plt.plot(x,y, 'x-', color='red', alpha=.5, label='the data')

plt.title('The Title')
plt.xlabel('x axis label')
plt.ylabel('y axis label')

plt.legend(loc='best') #loc = 'best' makes the legend not overlap the data

plt.grid()

plt.savefig('test_fig.png')
````

````python
# linijski grafici i histogram
t=np.linspace(1,100,100)
x1=np.sin(20./(2*np.pi) * t)
x2=np.sin(40./(2*np.pi)*t)
x3=x1+x2

plt.figure()
plt.subplot(2,1,1)
plt.plot(t,x1,label='x1')
plt.plot(t,x2,label='x2')
plt.plot(t,x3,label='x3')
plt.legend(loc='best')

plt.subplot(2,1,2)
x1hist=plt.hist(x1,density='true', bins=25, alpha=.25,label='x1')
x1hist=plt.hist(x2,density='true', bins=25, alpha=.25,label='x2')
plt.legend(loc='best')
````
